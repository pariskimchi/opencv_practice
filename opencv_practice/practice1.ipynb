```python
import cv2
import numpy as np
```

```python
# 영상 생성 

# 검은 바탕 이미지
img1 = np.empty((200,300), dtype=np.uint8)


cv2.imshow('img1',img1)

cv2.waitKey(0) #키보드 입력 대기, 입력한 키값 반환
cv2.destroyAllWindows() # 모든 윈도우 창 닫음
```

```python
img2=np.zeros((200,300,3),dtype=np.uint8) # 칼라 이미지

cv2.imshow('img2',img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 칼라 이미지 노란바탕
img4 = np.full((200,300,3),(0,255,255),dtype=np.uint8) 

cv2.imshow('img4',img4)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 읽기 
# cv2.imread('파일경로',flag)

# cv2.IMREAD_COLOR 칼라이미지. 상수1
fname = 'a.jpg'

original = cv2.imread(fname, cv2.IMREAD_COLOR)
gray = cv2.imread(fname,cv2.IMREAD_GRAYSCALE)
unchange = cv2.imread(fname, cv2.IMREAD_UNCHANGED)

print(original.shape)
print(gray.shape)
print(unchange.shape)

cv2.imshow('Original',original)
cv2.imshow('Gray',gray)
cv2.imshow('Unchange',unchange)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 이미지 쓰기 
# cv2.imwrite('파일명',img)
# img-저장할 이미지값 (numpy 배열)


fname_o = 'a.jpg'
fname_new ='a_gray.jpg'

gray = cv2.imread(fname_o, cv2.IMREAD_GRAYSCALE)

# 이미지 쓰기 
cv2.imwrite(fname_new, gray)

img = cv2.imread(fname_new)

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Matplotlib 에 이미지 출력

보통 칼라 표현 [r,g,b]로 표현, 
opencv는 [b,g,r]로 표현한다 
그래서 opencv로 읽어들인 이미지를 다른 프로그램을 통해 출력하면 
이미지 색상이 반전되어 출력되므로 출력하기전 r,g,b로 반전

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('a.jpg',cv2.IMREAD_COLOR)

plt.imshow(img)

plt.show()
```

```python
# 색상이 반전되었으므로, 원래색상대로 출력하려면 
# b,g,r => r,g,b

img = cv2.imread('a.jpg',cv2.IMREAD_COLOR)

b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])

plt.imshow(img2)
plt.xticks([])
plt.yticks([])
plt.show()
```
